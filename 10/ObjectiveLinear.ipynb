```julia
using LinearAlgebra
using PyPlot
```

```julia
G = [1 -1;2 -1;1 1];
d = [-1; 0; 2.5];
```

```julia
Phi = zeros(200, 400)
for i1 in 1:200
    m1 = i1 / 100
    for i2 = 1:400
        m2 = i2 / 100
        Phi[i1, i2] = norm(G*[m1, m2] - d)
    end
end
matshow(transpose(Phi));
colorbar()
set_cmap("Spectral_r")
xlabel("m1 * 100")
ylabel("m1 * 100")
```

```julia
plot(Phi[80, :])
plot(Phi[:, 160])
```
